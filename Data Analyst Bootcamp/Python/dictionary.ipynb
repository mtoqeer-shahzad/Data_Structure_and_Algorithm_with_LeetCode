```python
#create a list to store and calculate the average students grades

grades=[60,70,80,75,90]
average_grade=sum(grades)/len(grades)

print(f"Average Grade:{average_grade:.2f}")

#find Min and highest number

min_num=min(grades)
max_num=max(grades)

print("min_value",min_num)
print("max_num",max_num)
```

# create a list to collect and analyze the feedback

```python
feedback_list=['So interested','Helpful','partially']
feedback_list.append('Good job')

positive_feed_back=sum(1 for comment in feedback_list if 'helpful' in comment.lower() or 'so' in comment.lower())

print(f"Positive:{positive_feed_back}")


for comments in feedback_list:
    
    print(f"{comments}")
```

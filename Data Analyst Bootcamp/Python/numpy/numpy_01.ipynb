# SQL Lite

```python
import sqlite3
```

```python
connection=sqlite3.connect("example.db")
```

```python
cursor=connection.cursor()
```

```python
cursor.execute("""
    CREATE TABLE IF NOT EXISTS Employee (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        Age INTEGER
    )
""")

connection.commit()
```

```python
cursor.execute("INSERT INTO Employee(name, Age) VALUES(?,?)",("TOQEER",25))
connection.commit()
```

```python
cursor.execute("SELECT * FROM Employee")
```

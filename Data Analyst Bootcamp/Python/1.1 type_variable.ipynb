```python
def generate_square(n):
   square=[]
   
   for i in range(n):
       square.append('*'*n)
   return square
    
generate_square(2)
```

```python
def cal_sum_nat():
    n = 1        # start from 1
    total = 0    # store sum
    
    while n <= 10:
        total += n   # total = total + n
        n += 1       # increment n
    
    print("Sum of first 10 Natural Numbers:", total)
    print("number of first 10 Natural Numbers:", n)
    

cal_sum_nat()
```

```python
#calculate sum of first N natural Number using for loop
def cal_sum_N():
    n = 10
    total = 0
    
    for i in range(1, n+1):
        total =total + i  # total = total + i
    
    print("Sum of first", n, "Natural Numbers:", total)

cal_sum_N()
```

```python
def add_sum(nums,target):
    map_dict={}
    
    for i,n in enumerate(nums):
        diff=target-n
        
        if diff in map_dict:
            return[map_dict[diff],i]
    
        map_dict[n]=i
        
    return []


my_nums = [2, 7, 11, 15]
my_target = 9

result = add_sum(my_nums, my_target)
        
for row in result:
    print(row)
        
```

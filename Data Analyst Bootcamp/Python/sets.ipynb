```python
#union opeartion
set1={1,2,3,4,5,6,7,8}
set2={3,4,5,5,6,7,9}
union=set1.union(set2)
print(union)

#intersection

intersection=set1.intersection(set2)

print(intersection)


#differenece

diff=set1.difference(set2)
print(diff)


#syymetric diffence

s_d=set1.symmetric_difference(set2)
print(s_d)

#check subset
set_1={1,2,3}
set_2={3,4,5}
sb_set=set_1.issubset(set_2)
print(sb_set)

```

```python
unique_text='How to get unique word in text'
word=unique_text.split()
#convert list of word to set to get unique words

unique_word=set(word)
print(unique_word)
print(len(unique_word))
```
